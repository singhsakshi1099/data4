```python
from matplotlib import pyplot as plt
import numpy as np 

%matplotlib inline

a = np.arange(0, 10)
b = a * 2
c = a * 3
d = a * 4
e = a * 5
f = a * 6
g = a * 7
h = a * 8
i = a * 9
j = a * 10

fig  = plt.figure()
ax1 = fig.add_axes([0, 0 ,1 , 1])
ax2 = fig.add_axes([0.25,0.65,0.3,0.3])
ax1.plot(a, a, label="x")
ax1.plot(a, b, label="x2")
ax1.plot(a, c, label="x3")
ax1.plot(a, d, label="x4")
ax1.plot(a,e, label="x5")
ax1.plot(a,f,label="x6")
ax1.plot(a,g,label="x7")
ax1.plot(a,h,label="x8")
ax1.plot(a,i,label="x9")
ax1.plot(a,j,label="x10")

ax2.plot(a, a, label="x")
ax2.plot(a, b, label="x2")
ax2.plot(a, c, label="x3")
ax2.plot(a, d, label="x4")
ax2.plot(a,e, label="x5")
ax2.plot(a,f,label="x6")
ax2.plot(a,g,label="x7")
ax2.plot(a,h,label="x8")
ax2.plot(a,i,label="x9")
ax2.plot(a,j,label="x10")

ax2.set_xlim([0, 1])
ax2.set_ylim([0 ,7.5])
ax1.set_xlabel("Value of X")
ax1.set_xlabel("result of X times 1-10" )
ax1.set_title("value of 0-9 times 1-10")

ax2.set_title("zoom on values of 0 and 1")

ax1.legend()
```
